```python
import math
import sympy
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
plt.style.use('seaborn')

def get_coordinate(num):
    return num * np.cos(num), num * np.sin(num)

def create_plot(nums, figsize=8, s=None, show_annot=False):
    nums = np.array(list(nums))
    x, y = get_coordinate(nums)
    plt.figure(figsize=(figsize, figsize))
    plt.axis("off")
    plt.scatter(x, y, s=s)
    plt.savefig("primesplot.png", dpi=200)
    
primes = sympy.primerange(0, 10000)
create_plot(primes)
```
